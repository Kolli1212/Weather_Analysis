```python
data_clean.isnull().sum()  #sum function is use of True and False value 
```

```python
#fill No--> fillna, ffill,bfill
# MOCT
```

```python
x= round(data_clean.Ozone.mean())     #Round function is use for decimal value for converting int value  
x
```

```python
y = data_clean.Solar.median()        #median and mean use for numrical value
y
```

```python
z = data_clean.Weather.mode()        #mode use for Category
z
```

```python
z = data_clean.Weather.mode()[0]
z
```

```python
data_clean["Ozone"] = data_clean["Ozone"].fillna(x)    #fillna function use for filling nullvalue
```

```python
data_clean["Solar"] = data_clean["Solar"].fillna(y)
```

```python
data_clean["Weather"] = data_clean["Weather"].fillna(z)
```

```python
data_clean["Weather"] = data_clean["Weather"].ffill()    #here useing forward fill function
```
